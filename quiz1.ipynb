```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def logistic(x, L, k, x0):
    y = L / (1 + np.exp( -k * (x - x0) ))
    return y
```

```python
ngridpoints = 500 # you may decrease or increase this number
xgridpoints = np.linspace(-6, 6, ngridpoints)
```

```python
f0 = logistic(xgridpoints, L = 1, k = 1, x0 = 0)
f1 = logistic(xgridpoints, L = 1, k = 1, x0 = 2)
f2 = logistic(xgridpoints, L = 1, k = 2, x0 = 0)
```

```python
plt.plot(xgridpoints, f0)
plt.plot(xgridpoints, f1)
plt.plot(xgridpoints, f2)
```

```python
parameter = [
    [1, 1, 0],
    [1, 1, 2],
    [1, 2, 0]
]
flist = []
for ps in parameter:
    L = ps[0]
    k = ps[1]
    x0 = ps[2]
    f = logistic(xgridpoints, L, k , x0)
    flist.append(f)
    plt.plot(xgridpoints, f, label = f"L:(L)  k:(k)  x0:(x0)")
plt.grid(True)
plt.legend()
```

```python
parameters = [ (1, 1, 0), (1, 1, 2), (1, 2, 0)]

for L, k, x0 in parameters:
    f = logistic(x = xgridpoints, L=L, k=k, x0=x0)
    plt.plot(xgridpoints, f)
```
