```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def lineEquation(x, a, b):
    y = a * x + b
    return y
```

```python
y = lineEquation(x=1, a=2, b=3)
print(y)
```

```python
a = 2
b = 3

xvalues = [-1, 0, 1, 2, 3]
yvalues = []
for x in xvalues:
    y = lineEquation(x, a, b)
    yvalues.append(y)
```

```python
yvalues
```

```python
plt.plot(xvalues, yvalues, 'o-')
```

```python
# y = a * x^2 + b * x + c
# a = 1, b = 2, c = 3
# draw x-y plot for x in [-2, 4]
```

```python
def lineEquation2(x, a, b, c):
    y = a * x**2 + b * x + c
    return y

a = 1
b = 2
c = 3
xvalues2 = np.linspace(-2, 4, 100)
yvalues2 = []
for x in xvalues2:
    y = lineEquation2(x, a, b, c)
    yvalues2.append(y)

plt.plot(xvalues2, yvalues2, '-')
```
