```python
x = np.linspace(0, 10, 101)
print(x)
```

```python
y_cos = np.cos(x)
y_cos
```

```python
plt.plot(x, y_sin, label="sin")
plt.plot(x, y_cos, "--", label="cos")
plt.legend()
plt. title("Sin Cos")
plt.xlabel("xlable")
plt.ylabel("ylable")

plt.savefig("myFigure.pdf")
```
